```python
#importing frameworks
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime as dt

```

```python
#IMPORTING THE DATASETS:

covid_data=pd.read_csv('covid_19_india.csv') #reading the first dataset
covid_data.head(10)
```

```python
covid_data.describe() #to check the statistical summary of the data
```

```python
vaccine_data=pd.read_csv('covid_vaccine_statewise.csv') #reading the second dataset
vaccine_data.head(10) 
```

```python
vaccine_data.info() #to check the data type of each column
```

```python
vaccine_data.describe() #to check the statistical summary of the data
```

```python
#DATA CLEANING:
#1. covid_data
covid_data.isnull().sum() #to check the null values in the dataset
covid_data.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],axis=1,inplace=True) #dropping the unnecessary columns permanently
covid_data.head(10)
```

```python
#active cases:

covid_data["Active"]=covid_data["Confirmed"]-covid_data["Cured"]-covid_data["Deaths"]
covid_data.tail()
```

```python
#pivot table to check the total cases statewise:
state=pd.pivot_table(covid_data,values=["Confirmed","Deaths","Cured"],
                 index="State/UnionTerritory",aggfunc="max")
state
```

```python
#recovery rate:
state["recovery_rate"]=state["Cured"]/state["Confirmed"]*100
state.head(10)
```

```python
#death rate:
state["death_rate"]=state["Deaths"]/state["Confirmed"]*100
state.tail(10)
```

```python
state=state.sort_values(by="Confirmed",ascending=False) #sorting the values in descending order

state.style.background_gradient(cmap="cool") #to show the data in the form of heatmap
```

```python
#top 20 active cases states:

top20=covid_data.groupby(by="State/UnionTerritory")["Active"].sum().sort_values(ascending=False).head(20)
top20=pd.DataFrame(top20)
top20
```

```python
top_20=sns.barplot(x=top20["Active"].head(20),y=top20.index[:20],
                   edgecolor="green",linewidth=3,data=state);
plt.title("Top 20 states with most active cases",size=20)
plt.xlabel("Active cases",size=15)
plt.ylabel("States",size=20)
```

```python
#top 20 states with highest death cases:

top20_d=covid_data.groupby(by="State/UnionTerritory")["Deaths"].sum().sort_values(ascending=False).head(20)
top20_d=pd.DataFrame(top20_d)
top20_d
```

```python
sns.barplot(x=top20_d["Deaths"].head(20),y=top20_d.index[:20]);
plt.title("Top 20 states with most death cases",size=20)
plt.xlabel("States",size=15)
plt.ylabel("Death cases",size=20)
```

```python
vaccine_data.isnull().sum() #to check the null values in the dataset
```

```python
#dropping the columns that have more than 70% null values:

vaccine_data.drop(vaccine_data.columns[vaccine_data.isnull().sum()>0.7*len(vaccine_data)],axis=1,inplace=True)
vaccine_data.isnull().sum()
```

```python
#rename the columns:
vaccine_data.rename(columns={"Updated On":"Date","State/UnionTerritory":"State"},inplace=True)
vaccine_data.head()
```

```python
#male and female vaccinated:

ml=vaccine_data["Male (Doses Administered)"].sum()
fm=vaccine_data["Female (Doses Administered)"].sum()

px.pie(names=["Male","Female"],values=[ml,fm],title="Male and Female vaccinated")
```
